```python
import os
import yaml

# Get the user's home directory
home_dir = os.path.expanduser('~/Projects/NCCU_GradProj')

# Define data
data = {
    'train': os.path.join(home_dir, 'data/train/'),
    'val': os.path.join(home_dir, 'data/val/'),
    'test': os.path.join(home_dir, 'data/test/'),
    'names': ['people'],
    'nc': 1, # number of classes
}

# File path
file_path = os.path.join(home_dir, 'check.yaml')
directory = os.path.dirname(file_path)

# Create YAML file
with open(file_path, 'w') as f:
    yaml.dump(data, f)

# Read YAML file
with open(file_path, 'r') as f:
    taco_yaml = yaml.safe_load(f)
    print(taco_yaml)
```

```python
from ultralytics import YOLO
import os

# Modify to use absolute path
yaml_path = os.path.join(home_dir, 'check.yaml')
model_path = os.path.join(home_dir, 'models/yolov8n.pt')
proj_path = os.path.join(home_dir, 'train_info')

# Load model and train
model = YOLO(model_path)
model.train(data=yaml_path, epochs=15, batch=32, imgsz=416, project=proj_path)
```

```python
trained_model_path = os.path.join(home_dir, 'models/best.pt')  # Path to the best model weights

# Load the trained model
model = YOLO(trained_model_path)

# Evaluate the model on the test dataset
results = model.val(data=yaml_path)
```
